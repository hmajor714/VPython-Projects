```python
# ** GLOBAL VARIABLES ** #
side = 4.0
thk = 0.3
s2 = 2*side - thk
s3 = 2*side + thk

# Floors Declaration
floors = [1, 2, 3, 4]

# *** OBJECT SETUP *** # (ex. elevator, shaft, text, etc.)
# ** ELEVATORS ** #
class elevator:
    queue = []
    opdirqueue = []
    direction = 0 #0 at rest, 1 up, 2 down
    def sortQ(self): # just insertion sort
        #print("queue before: ", self.queue)
        if (self.direction == 1):
            #print("Direction is up")
            for i in range(1, len(self.queue)):
                temp = self.queue[i]
                j = i-1
                while j >= 0 and temp < self.queue[j] :
                        self.queue[j + 1] = self.queue[j]
                        j -= 1
                self.queue[j + 1] = temp
        if (self.direction == 2):
            #print("direction is down")
            for i in range(1, len(self.queue)):
                temp = self.queue[i]
                j = i-1
                while j >= 0 and temp > self.queue[j] :
                        self.queue[j + 1] = self.queue[j]
                        j -= 1
                self.queue[j + 1] = temp
        print("queue: ", self.queue)
        print("opdirqueue: ", self.opdirqueue)
    def __init__(self):
        # Creating the walls of the elevator unit
        self.elevator_right = box (pos=vector( side, 0, 0), size=vector(thk, s2, s3),  color = color.black)
        self.elevator_left = box (pos=vector(-side, 0, 0), size=vector(thk, s2, s3),  color = color.black)
        self.elevator_bottom = box (pos=vector(0, -side, 0), size=vector(s3, thk, s3),  color = color.black)
        self.elevator_top = box (pos=vector(0,  side, 0), size=vector(s3, thk, s3),  color = color.black)
        self.elevator_back = box(pos=vector(0, 0, -side), size=vector(s2, s2, thk), color = color.black)
        self.current = 2

        # Creating the doors of the elevator unit
        self.door_left = box (pos=vector(side/2, 0, side), size=vector(s2/2, s2, thk), color = color.gray(0.7))
        self.door_right = box (pos=vector(-side/2, 0, side), size=vector(s2/2, s2, thk), color = color.gray(0.7))

        # Creating a compound of the shapes
    def compound(self): 
        self.doors = compound([self.door_left, self.door_right], pos=vector(0,0,0))
        self.unit = compound([self.elevator_right, self.elevator_left, self.elevator_bottom, self.elevator_top, self.elevator_back], pos=vector(0,0,0))
        
        
# inherited class from elevator
class f_elevator(elevator):
    def __init__(self):
        super().__init__()
        self.elevator_back.opacity = 0.0
        self.b_door_left = box (pos=vector(side/2, 0, side), size=vector(s2/2, s2, thk), color = color.gray(0.7))
        self.b_door_right = box (pos=vector(-side/2, 0, side), size=vector(s2/2, s2, thk), color = color.gray(0.7))
        self.b_doors = compound([self.b_door_left, self.b_door_right], pos=vector(0,0,0))

# ** SHAFTS ** #
class shaft:
    def __init__(self):
        self.shaft_right = box (pos=vector( side*1.1, 0, 0), size=vector(thk, s2*4.18, s3*1.1),  color = color.white, opacity=0.5)
        self.shaft_left = box (pos=vector(-side*1.1, 0, 0), size=vector(thk, s2*4.18, s3*1.1),  color = color.white, opacity=0.5)
        self.shaft_bottom = box (pos=vector(0, -side*4, 0), size=vector(s3, thk, s3*1.1),  color = color.white, opacity=0.5)
        self.shaft_top = box (pos=vector(0,  side*4, 0), size=vector(s3, thk, s3*1.1),  color = color.white, opacity=0.5)
        self.shaft_back = box(pos=vector(0, 0, -side*1.1), size=vector(s2*1.18, s2*4.18, thk), color = color.white, opacity=0.5)

    def compound(self): 
        self.unit = compound([self.shaft_right,self.shaft_left,self.shaft_bottom,self.shaft_top,self.shaft_back], pos=vector(0,0,0))
        
# button class
class button():
    def turnOn(self):
        rate(100)
        self.orb.color = vector(1,1,0)
    def turnOff(self):
        rate(100)
        self.orb.color = vector(1,1,1)
    def __init__(self, floor, posx, posy, direction = None):
        self.floor = floor
        self.posx = posx
        self.posy = posy
        self.isOn = False
        self.orb = sphere(pos=vector(self.posx,self.posy,5),radius=2,color=vector(1,1,1))
        if (direction is not None):
            self.direction = direction
            if (self.direction == 1):
                self.txt = text(pos = vector(self.posx, self.posy, 8),text="^", align = 'center', color = vector(1,0,0))
            elif (self.direction == 2):
                self.txt = text(pos = vector(self.posx, self.posy, 8),text="v", align = 'center', color = vector(1,0,0))
        else:
            self.txt = text(pos = vector(self.posx, self.posy-0.5, 8),text=str(self.floor), align = 'center', color = vector(1,0,0))
        #self.unit = compound([self.orb, self.txt], pos = vector(0,0,0))

# ** BOILERPLATE SETUP ** #
# Pass in the elevator object, shaft object, text name, and the x-position
def create_scene(elevator, name_shaft, name, position):    
    
    elevator.compound()
    title = text(text=name, pos=vector(position, 20, 0), align="center", height=2)
    name_shaft.compound()
    elevator.unit.pos = vector(position,-4,0)
    elevator.doors.pos = vector(position,-4,side)
    name_shaft.unit.pos = vector(position,0,0)
    
    #if name == "FREIGHT":
        #elevator.b_doors.pos = vector(position,-4,-side)
    
    # !!! Do we need this? not sure how reference vs value pointer stuff works in Python
    return elevator

def doorOpenClose(elevator):
    #need to check if this is freight if it is skip this code and do new code for two doors
   # if(freight):
        #print("BACK DOOR OPENING")
    while(elevator.doors.opacity>0):
        rate(10)
        elevator.doors.opacity=elevator.doors.opacity-0.05
    while(elevator.doors.opacity<1):
        rate(10)
        elevator.doors.opacity=elevator.doors.opacity+0.05
    return
        

def moveUp(elevator, current):
    dest = current + 1
    for i in range(0,80):
        rate(10)
        elevator.unit.pos = elevator.unit.pos + vector(0,0.1,0)
        elevator.doors.pos = elevator.doors.pos + vector(0,0.1,0)
        i = i + 1
    sleep(1)
    return dest
        
def moveDown(elevator, current):
    dest = current - 1 
    for i in range(0,80):
        rate(10)
        elevator.unit.pos = elevator.unit.pos - vector(0,0.1,0)
        elevator.doors.pos = elevator.doors.pos - vector(0,0.1,0)
        i = i + 1
    sleep(1)
    return dest

def ele_to_you(elevator, which_floor):
    while(elevator.current != which_floor):
        if(elevator.current >  which_floor):
            elevator.current = moveDown(elevator, elevator.current)
        elif(elevator.current <  which_floor):
            elevator.current = moveUp(elevator, elevator.current)
    return

# ****** START OF MAIN ****** #
def main():
    #scene.canvas()
    scene.caption = """<h3 style="line-height:0px;">
                    Elevator Project</h3>
                    - by Alex, David, Grace, Hunter, Lleyton
                    ...comments go here..."""
    scene.userzoom = False
    
    # boilerplate✨✨✨✨
    west_elevator = elevator()
    west_shaft = shaft()
    create_scene(west_elevator, west_shaft, "WEST", -15)
    
    #freight_elevator = f_elevator()
    #freight_shaft = f_shaft()
    freight_elevator = elevator()
    freight_shaft = shaft()
    create_scene(freight_elevator, freight_shaft, "FREIGHT", 0)
    
    east_elevator = elevator()
    east_shaft = shaft()
    create_scene(east_elevator, east_shaft, "EAST", 15)
    
    elevs = { # Elevator dictionary
        1:west_elevator,
        2:freight_elevator,
        3:east_elevator
    }
    but1 = button(1, -25, -8)
    but1r = button("1R", -30, -8)
    but2 = button(2, -25, -3)
    but3 = button(3, -25, 2)
    but4 = button(4, -25, 7)
    but4r = button("4R", -30, 7)
    
    buts = {
        1:but1,
        2:but2,
        3:but3,
        4:but4
    }
    
    # Start of logic 
    sleep(2)
    which_ele = input("Please enter an elevator to use (1-3): ")
    while int(which_ele) > 4 or int(which_ele) < 1:
        print("Please input a valid elevator")
        which_ele = input("Please enter an elevator to use (1-3): ")
        
    which_flo = input("Which floor are you on (1-4): ")
    while int(which_flo) > 4 or int(which_flo) < 1:
        print("Please input a valid floor")
        which_flo = input("Which floor are you on (1-4): ")
        
    which_dir = input("Please enter 1 for up and 2 for down: ")
    while int(which_dir) > 2 or int(which_dir) < 1:
        print("Please input a valid int")
        which_dir = input("Please enter 1 for up and 2 for down: ")

    which_elevator = int(which_ele)
    which_floor = int(which_flo)
    which_direction = int(which_dir)
    elevs[which_elevator].direction = which_direction

    # while
    stop = False
    while(stop == False):
        rate(100)
        if(which_elevator == 1):
            ele_to_you(west_elevator, which_floor)
        elif(which_elevator == 2):
            ele_to_you(freight_elevator, which_floor)
        elif(which_elevator == 3):
            ele_to_you(east_elevator, which_floor)
        stop = True
    if(which_elevator == 1):        
        doorOpenClose(west_elevator)
    elif(which_elevator == 2):        
        doorOpenClose(freight_elevator)
    elif(which_elevator == 3):        
        doorOpenClose(east_elevator)
    
    sleep(2)
    
    which_dest = input("Please enter a floor to travel to (1-4): ")
    while int(which_dest) > 4 or int(which_dest) < 1:
        print("Please input a valid floor")
        which_dest = input("Please enter a floor to travel to (1-4): ")
        
    which_destination = int(which_dest)
    #light on
    buts[which_destination].turnOn()
    #add destination to the queue
    elevs[which_elevator].queue.append(which_destination)    
    
    #handle freight elevator
    if(which_elevator == 2):
        if(which_destination == 4 or which_destination == 1):
            bd = input("Did you enter the elevator through the back door (1-yes, 2-no) ?")
            while int(bd) > 2 or int(bd) < 1:
                print("Please input a valid int")
                bd = input("Did you enter the elevator through the back door (1-yes, 2-no) ?")
            back_door = int(bd)
            if(back_door==1):
                key_c = input("Do you have the proper key card (1-yes, 2-no) ?")
                while int(key_c) > 2 or int(key_c) < 1:
                    print("Please input a valid int")
                    key_c = input("Do you have the proper key card (1-yes, 2-no) ?")
                key_check = int(key_c)
                if(key_check==1):
                    if which_destination == 4:
                        buts[4].turnOff()
                        but4r.turnOn()
                    elif which_destination == 1:
                        buts[1].turnOff()
                        but1r.turnOn()
                    print("Traveling uninterrupted to floor")
                    ele_to_you(freight_elevator, which_destination)
    else:

        #while the queue is not empty   
        while(len(elevs[which_elevator].queue) > 0):
            more_floors = True
            while(more_floors == True):
                which_dest = input("Would you like to add another floor to the queue (1-4)? Enter '0' if no: ")
                if(int(which_dest) == 0):
                    more_floors = False
                    
               # while int(which_dest) > 4 or int(which_dest) < 1:
                #    print("Please input a valid floor")
                 #   which_dest = input("Please enter an floor to go to (1-4): ")
                
                    
                which_destination = int(which_dest)
                #elevs[which_elevator].sortQ()
                if(which_destination >= 1 and which_destination <= 4):
                    #checking if its an opposite direction
                    if(elevs[which_elevator].direction == 1 and which_destination < elevs[which_elevator].current):
                        elevs[which_elevator].opdirqueue.append(which_destination)
                    elif(elevs[which_elevator].direction == 2 and which_destination > elevs[which_elevator].current):
                        elevs[which_elevator].opdirqueue.append(which_destination)
                    else:
                        #turn light on
                        buts[which_destination].turnOn()
                        #add destination to the queue
                        elevs[which_elevator].queue.append(which_destination)
                    #sort queue
                    elevs[which_elevator].sortQ()


            #the first floor to travel in the queue is above us
            if(elevs[which_elevator].queue[0] > elevs[which_elevator].current):
                #move up 1 floor
                elevs[which_elevator].current = moveUp(elevs[which_elevator], elevs[which_elevator].current)
                #is this where we are supposed to stop?
                if(elevs[which_elevator].current == elevs[which_elevator].queue[0]):
                    doorOpenClose(elevs[which_elevator])
                    #turn light off
                    buts[elevs[which_elevator].queue[0]].turnOff()
                    #remove floor from queue
                    elevs[which_elevator].queue.pop(0)
            #the first floor to travel in the queue is below us
            elif(elevs[which_elevator].queue[0] < elevs[which_elevator].current):
                elevs[which_elevator].current = moveDown(elevs[which_elevator], elevs[which_elevator].current)
                #is this where we are supposed to stop?
                if(elevs[which_elevator].current == elevs[which_elevator].queue[0]):
                    doorOpenClose(elevs[which_elevator])
                    #turn light off
                    buts[elevs[which_elevator].queue[0]].turnOff()
                    #remove floor from queue
                    elevs[which_elevator].queue.pop(0)
            # checking if queue is finished and if there are any floors in the opposite direction
            if(len(elevs[which_elevator].opdirqueue) > 0 and len(elevs[which_elevator].queue) == 0):
                #adding the opdirqueue to the regular queue
                elevs[which_elevator].queue.extend(elevs[which_elevator].opdirqueue)
                #Swapping directions as needed
                if(elevs[which_elevator].direction == 2):
                    elevs[which_elevator].direction = 1
                elif(elevs[which_elevator].direction == 1):
                    elevs[which_elevator].direction = 2 
                elevs[which_elevator].sortQ()
                elevs[which_elevator].opdirqueue.clear()
    print("There are no more floors left to travel")
    but4r.turnOff()
    but1r.turnOff()
            
        
if __name__ == "__main__":
    main()
```
